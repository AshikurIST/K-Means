```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

#  K means clustaring work with unsupervised data

Import data from sns

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0]
```

```python
data[1]       # correct label
```

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

# K Means Clustaring algorithm wiht sklearn

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])          # fit the freature data
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_         # predictade label
```

```python
fig, (ax1,ax2)=plt.subplots(1,2, sharey=True,figsize=(10,6))    # in subplot 1,2 is one row by two columns

ax1.set_title('k means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
